```python
import cv2
from IPython.display import display, Image

#pre-trained face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

#read image
input_image_path = '/content/new/1000107352.jpg'  # Provide the correct path to your PNG input image
image = cv2.imread(input_image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Detect faces in the image
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

#Draw rectangles/bounding box around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

#Saving the output
output_image_path = '/content/new/1000107352_transformed.jpg'
cv2.imwrite(output_image_path, image)

# Display image (using IPython display module)
display(Image(filename=output_image_path))
```
